```python
import gudhi
from petls import Complex
import tadasets
import math
```

```python
npoints = 150 # works on up to ~150
torus = tadasets.torus(n=npoints, c= 2, a=0.5, noise = 0)
# torus = [[-1.94383418,  0.21319095,  0.99900895], #row 1
#  [ 1.46405456, -2.61071902,  0.11633033], # row 2
#  [-1.3485262,  -0.44939415,  0.81563642], # row 3
#  [ 2.35950996, -1.1217201,   0.79041319], # row 4
#  [-0.41958451, -0.94801706, -0.26849738] # row 5
#  ]

torus2 = tadasets.torus(n=npoints, c= 2, a=1, noise = 0)
# torus2 = [[-1.94383418,  0.21319095,  0.99900895], #row 1
#  [ 2.35950996, -1.1217201,   0.79041319], # row 4
#  [-0.41958451, -0.94801706, -0.26849738], # row 5
#  [ 1.46405456, -2.61071902,  0.11633033], # row 2
#  [-1.3485262,  -0.44939415,  0.81563642] # row 3
#  ]


print(torus)
scale = 0.1
max_filt = 3
dim = 0
num_steps = math.ceil(max_filt/scale)-1
samples = [x*scale for x in range(num_steps+1)]
spectra_request = [(dim,samples[i],samples[i+1]) for i in range(len(samples)-1)]
```

```python
%matplotlib ipympl
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(torus[:,0],torus[:,1],torus[:,2])
plt.show()
```

```python
rips_complex = gudhi.RipsComplex(points=torus,
                                 max_edge_length=max_filt)

simplex_tree = rips_complex.create_simplex_tree(max_dimension=2)
result_str = 'Rips complex is of dimension ' + repr(simplex_tree.dimension()) + ' - ' + \
    repr(simplex_tree.num_simplices()) + ' simplices - ' + \
    repr(simplex_tree.num_vertices()) + ' vertices.'
print(result_str)
fmt = '%s -> %.2f'
for filtered_value in simplex_tree.get_filtration():
    print(fmt % tuple(filtered_value))

rips_complex2 = gudhi.RipsComplex(points=torus2,
                                 max_edge_length=max_filt)

simplex_tree2 = rips_complex2.create_simplex_tree(max_dimension=2)
# result_str = 'Rips complex is of dimension ' + repr(simplex_tree2.dimension()) + ' - ' + \
#     repr(simplex_tree2.num_simplices()) + ' simplices - ' + \
#     repr(simplex_tree2.num_vertices()) + ' vertices.'
# print(result_str)
# fmt = '%s -> %.2f'
# for filtered_value in simplex_tree2.get_filtration():
#     print(fmt % tuple(filtered_value))
```

```python
pl = Complex(simplex_tree=simplex_tree)
pl.verbose = True
```

```python
st = pl.spectra(request_list=spectra_request)
print(st)
```

```python
pl2 = Complex(simplex_tree=simplex_tree2)
pl2.verbose = True
st2 = pl2.spectra(request_list=spectra_request)
```

```python
st
ripser

def get_graph(spectrum, dim):
    graph = []
    for s in spectrum:
        if s[0] == dim:
            a = s[1]
            eigs = s[3] 
            nonzero = [x for x in eigs if x> 0.1]
            if len(nonzero) > 0:
                mineig = min(nonzero)
            else:
                mineig = 0
            graph.append((a, mineig))
    return graph

stgraph = get_graph(st, dim)
rgraph  = get_graph(ripser, dim)

stgraph2 = get_graph(st2, dim)
rgraph2  = get_graph(ripser2, dim)
# stgraph = [(s[1], min([x for x in s[3] if x != 0])) for s in st if s[0]==dim and ]
print(stgraph)
print(rgraph)

lambda_s = [x[1] for x in stgraph]
lambda_r = [x[1] for x in rgraph]
lambda_s2 = [x[1] for x in stgraph2]
lambda_r2 = [x[1] for x in rgraph2]

a_vals = [x[0] for x in stgraph]
a_vals_check = [x[0] for x in rgraph]
if len(a_vals) != len(a_vals_check):
    print("len(a_vals) != len(a_vals_check)", len(a_vals), len(a_vals_check))
a_vals_diff = [abs(a_vals[i] - a_vals_check[i]) for i in range(len(a_vals))]
if max(a_vals_diff) > 1e-4:
    print("oh no")
    
```
